## Payment
- Preprocessing payment table
    - Choosing and 'Groupby' the PaymentStatus feature to get valuable insights
    - Save the new table as 'payment_processed.csv'

```python
raw_payment
```

```python
payment = raw_payment[['loanId', 'paymentStatus']].groupby(raw_payment['loanId']).describe()
payment
```

```python
payment_processed = payment.drop(columns=['loanId']).reset_index().drop(columns=('paymentStatus', 'unique'))
payment_processed
```

```python
payment_processed.to_csv('data/payment_processed.csv',index=False)
```
